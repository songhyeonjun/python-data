```python
import operator
from tkinter import *
import pymongo
import random



conn = pymongo.MongoClient() ## 몽고db서버에 연결
trip = conn.trip
place = trip.place

def random_list() : # DB에서 여행지 목록 랜덤으로 불러오기
    docs = place.find({"place_area": "제주도"}, {"_id": 0, "place_idx": 0})  # area가 제주도인것만 그리고 _id, Place_idx 뺴고
    docs_list_info = []
    for doc in docs:
        docs_list = list(doc.values())  # values값만 추출
        docs_list_info.append(docs_list[1])  # 리스트로 변환
    random_list_info = random.sample(docs_list_info, 10)  # 랜덤함수 사용 10개만 추출

    for i in random_list_info:
        Output.insert(END, i + "\n")

def like_list() : # 좋아요 높은 순으로 가져오기
    docs = place.find({}, {"_id": 0, "place_idx": 0}).sort("place_like", pymongo.DESCENDING)  # 내림차순으로 모든정보 가져오기
    docs_list_info = []
    for doc in docs:
        docs_list = doc.get("place_info")  # place_info만 추출
        docs_list_info.append(docs_list)  # list안에 넣기

    for i in docs_list_info[:5]: # 5개만 추출
        Output.insert(END, i + "\n")

def view_list() : # 조회수 높은 순으로 가져오기
    docs = place.find({}, {"_id": 0, "place_idx": 0}).sort("place_view", pymongo.DESCENDING)  # 내림차순으로 가져오기
    docs_list_info = []
    for doc in docs:
        docs_list = doc.get("place_info")  # place_info만 추출
        docs_list_info.append(docs_list)  # list안에 넣기

    for i in docs_list_info[:10]: # 10개만 추출
        Output.insert(END, i + "\n")





root = Tk()
root.geometry("1080x720")
root.title(" 배낭이 ")

def callback(event):
    Display.invoke() # 클릭 동작
    inputtxt.delete("1.0", "end")

def Take_input():
    INPUT = inputtxt.get("1.0", "end-1c")
    INPUT = INPUT.strip()
    if (INPUT == "추천") :
        Output.insert(END, INPUT.rjust(45))
        Output.insert(END, "\n")
        button1.pack()
        button2.pack()
    elif (INPUT == '랭킹조회') :
        Output.insert(END, INPUT.rjust(45))
        Output.insert(END, "\n")
        button3.pack()
        button4.pack()
    elif (INPUT == '일상대화') :
        pass
    else:
        Output.insert(END, INPUT.rjust(45))
        Output.insert(END, "\n")
        Output.insert(END, "공부하겠습니다.ㅜㅜ\n")
    inputtxt.delete("1.0", "end")

inputtxt = Text(root, height=3,
                width=50,
                bg="light yellow",
                )

Output = Text(root, height=40,
              width=50,
              bg="light cyan",
              )

Display = Button(root, height=2,
                 width=20,
                 text="Show",
                 command=lambda: Take_input())

button1 = Button(root, height=2,
               width=20,
               text="여행지 목록",
               command=lambda : random_list()
               )

button2 = Button(root, height=2,
               width=20,
               text="이미지로 추천 받기",

                 )

button3 = Button(root, height=2,
                width=20,
                text="조회수",
                command=lambda : view_list()

               )

button4 = Button(root, height=2,
               width=20,
               text="추천수",
               command=lambda : like_list()
                 )

Output.insert(INSERT,"안녕하세요 배낭이 입니다. 무엇을 도와드릴까요?\n")
Output.pack()
inputtxt.pack()
Display.place(x=718, y=525)

root.bind('<Return>', callback)
mainloop()

```
