```python
# x, y축으로 사용할 데이터를 주고, 1차방정식으로 표현했을 때
# 기울기와 y절편을 구해주세요. polyfit()
pf1 = np.polyfit(data_result['인구수'], data_result['소계'], 1)
pf1
```

```python
# 위에서 구한 기울기와 절편으로 함수를 만들어주세요
f1 = np.poly1d(pf1)
f1
```

```python
fx = np.linspace(100000, 700000, 100)
fx[:5]
```

```python
# 위에 그래프는 무엇을 나타내는가?
# 서울의 인구수에 따른 cctv 구별 설치 수를 나타낸다

# cctv를 어느 구에 설치해야하는가?
# 판단선 : 인구수에 적절한 cctv수를 판단하는 선
#          판단선 위에 위치한 구는 인구수에 비해 cctv가 많이 설치된 구
#          판단선 아래에 위치한 구는 인구수에 비해 cctv 설치가 부족한 구
#          판단선 아래에 있는 구에 cctv를 설치를 해야한다고 얘기할 수 있다.
#          오차가 큰 순으로! 먼저 선별을 하면 될 것 같다
# 위에 그래프를 바탕으로 도봉구 강북구 과진구, 등등이 있다..

# 판단기준 : 인구수와 cctv와 관련있다라는 기준
```
