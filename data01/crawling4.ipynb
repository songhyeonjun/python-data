```python
from urllib import request
```

```python
from bs4 import BeautifulSoup
```

```python
con = request.urlopen('https://novel.naver.com/best/ranking?genre=101&periodType=DAILY')
con 
```

```python
con.status #status code
```

```python
doc = BeautifulSoup(con, 'html.parser')
```

```python
li_tag_list = doc.select('#content > div > div.ranking_lst > div.ranking_wrap_right > ul > li')
li_tag_list
```

```python
tit_tag_list = doc.select('#content > div > div.ranking_lst > div.ranking_wrap_right > ul > li > a > div > p > span.tit')
tit_tag_list
```

```python
tit_collect = []
# for one in tit_tag_list:
#     tit_collect.append(one.text)
# tit_collect

for i in range(1, len(tit_tag_list)):
    tit1 = tit_tag_list[i].text.split(',')
    print(i, tit1)
    tit2 = tit1[0] + tit1[1]
    tit_collect.append(tit2)
tit_collect
```

```python
writer_tag_list = doc.select('#content > div > div.ranking_lst > div.ranking_wrap_right > ul > li > a > div > p > span.author')
writer_tag_list
```

```python
writer_collect = []
for one in writer_tag_list:
    writer_collect.append(one.text)
print(len(writer_collect))
```

```python
imsi1 = doc.select('#content > div > div.ranking_lst > div.ranking_wrap_right > ul > li > a > div > p > span.score_area')
imsi1
```

```python
lookup_collect = []
for one in imsi1:
    if len(one) == 2:
        lookup_collect.append(one.text)
lookup_collect
```

```python
img_tag_list = doc.select('#content > div > div.ranking_lst > div.ranking_wrap_right > ul > li > a > div.pic')
img_tag_list
```

```python
img_tag_collect = []
for li in img_tag_list :
    img = li.find('img')
    img_src = img['src']
    img_tag_collect.append(img_src)
img_tag_collect
```

```python
hottracts_file = open('romance.csv', 'a', encoding="UTF-8-sig")
#title1,brand1,percent1,price1
```

```python
for i in range(100):
    hottracts_file.write(img_tag_collect[i] + ',' +
                         tit_collect[i] + ',' +
                         writer_collect[i]  + ',' +
                         lookup_collect[i] + '\n')
```

```python
hottracts_file.close()
```
