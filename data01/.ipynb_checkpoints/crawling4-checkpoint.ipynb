```python
from urllib import request
```

```python
from bs4 import BeautifulSoup
```

```python
con = request.urlopen('https://novel.naver.com/best/ranking?genre=101&periodType=DAILY')
con 
```

```python
con.status #status code
```

```python
doc = BeautifulSoup(con, 'html.parser')
```

```python
li_tag_list = doc.select('#content > div > div.ranking_lst > div.ranking_wrap_right > ul > li')
li_tag_list
```

```python
tit_tag_list = doc.select('#content > div > div.ranking_lst > div.ranking_wrap_right > ul > li > a > div > p > span.tit')
tit_tag_list
```

```python
tit_collect = []
```

```python
tit_collect
for i in range(0, len(tit_tag_list)):
    tit_collect.append(tit_tag_list[i].text.replace(',',''))
tit_collect
```

```python
writer_tag_list = doc.select('#content > div > div.ranking_lst > div.ranking_wrap_right > ul > li > a > div > p > span.author')
writer_tag_list
```

```python
writer_collect = []
for one in writer_tag_list:
    writer_collect.append(one.text.replace(',',''))
writer_collect
```

```python
score = doc.select('#content > div > div.ranking_lst > div.ranking_wrap_right > ul > li > a > div > p > span.score_area')
score
```

```python
lookup_collect = []
for one in score:
    if len(one) == 2:
        lookup_collect.append(one.text.strip('별점'))
lookup_collect
```

```python
img_tag_list = doc.select('#content > div > div.ranking_lst > div.ranking_wrap_right > ul > li > a > div.pic')
img_tag_list
```

```python
img_tag_collect = []
for li in img_tag_list :
    img = li.find('img')
    img_src = img['src']
    img_tag_collect.append(img_src)
img_tag_collect
```

```python
hottracts_file = open('romance.csv', 'w', encoding="UTF-8-sig")
#title1,brand1,percent1,price1
```

```python
for i in range(100):
    hottracts_file.write(img_tag_collect[i] + ',' +
                         tit_collect[i] + ',' +
                         writer_collect[i]  + ',' +
                         lookup_collect[i] + '\n')
```

```python
hottracts_file.close()
```
