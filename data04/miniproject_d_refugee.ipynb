```python
col = ['망명년도', '망명지역', '실거주지', '난민 지위', '총 보류 누적 사람 수(시작 연도 기준으로 시작)', 
           '유엔난민기구 지원 사람 수(시작 연도 기준으로 시작)', '전쟁 난민 신청 수', '전쟁 난민 결정 수',
           '전쟁 난민 보류 수', '전쟁 난민 거부 수', '전쟁 난민 철회 수', '전쟁 난민 인원 현황',
          '총 보류 누적 사람 수(시작 연도 기준으로 종료)', '유엔난민기구 지원 사람 수(시작 연도 기준으로 종료)']
```

```python
refugee = pd.read_csv("./asylum_seekers.csv",header=1, low_memory=False, names=col)
```

```python
refugee = refugee.replace('*',0)
refugee2 = refugee.fillna(0)
refugee2 = refugee2.astype({'전쟁 난민 인원 현황':'int'})
refugee2
```

```python
refugee_decision = refugee2[['망명년도', '전쟁 난민 신청 수', '전쟁 난민 결정 수', '전쟁 난민 철회 수']]
refugee_decision = refugee_decision.astype({'전쟁 난민 신청 수':'int', '전쟁 난민 결정 수':'int', '전쟁 난민 철회 수':'int',})
refugee_decision
```

```python
decision_refugee1= refugee_decision.groupby('망명년도')['전쟁 난민 신청 수'].sum()
decision_refugee1 # type = Series
```

```python
decision_refugee2= refugee_decision.groupby('망명년도')['전쟁 난민 결정 수'].sum()
decision_refugee2 # type = Series
```

```python
decision_refugee3= refugee_decision.groupby('망명년도')['전쟁 난민 철회 수'].sum()
decision_refugee3 # type = Series
```

```python
total_d_refugee = pd.concat([decision_refugee1, decision_refugee2, decision_refugee3], axis=1)
total_d_refugee
```

```python
total_d_refugee.to_csv('total_d_refugee.csv', encoding='UTF-8-sig')
```
