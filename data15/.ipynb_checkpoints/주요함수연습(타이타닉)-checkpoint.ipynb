```python
## apply, map의 용도는 df의 컬럼의 값을 특정한 처리 후, df의 컬럼으로 다시 갱신하거나, 파생 변수를 만들어
## 집어넣기 위한 것.
## ==> apply, map의 함수로 사용되는 것은 입력값이 시리즈의 값 하나를 처리하는 단위로 만들어야 함
## ==> return을 반드시 넣어서 함수로 만들어야 한다.
## x는 시리즈(리스트)의 하나의 값을 처리하는 단위(값하나)
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../titanic_train.csv')
df
```

## data-set : titanic
### 1. p-class column update => 1 : low, 2-3 : middle, 4-5 : high
### 2. name으로 파생변수 name_result를 만들기<br>글자수를 세어 15이상이면 long, 나머지 short
### 3. survived 1이면 생존, 0이면 사망, survived column update
### 4. age 10대, 20대, 30대, 40대, 50대, 60대 이상으로 column update
### 5. name에 Mr, Dr를 포함하고 있으면 '남자', Miss를 포함하고 있으면 '여자', 나머지 '기타'<br>파생변수 name_result2를 만들고, 각각을 카운트 남자 몇명, 여자 몇명

### 1. p-class column update => 1 : low, 2-3 : middle, 4-5 : high

```python
def pclassupdate(x) :
    if x == 1 :
        return 'low'
    elif 2 <= x <=3:
        return 'middle'
    else :
        return 'high'
```

```python
df['pclass'] = df['pclass'].apply(pclassupdate)
```

### 2. name으로 파생변수 name_result를 만들기<br>글자수를 세어 15이상이면 long, 나머지 short

```python
df['name'].apply(lambda x : 'long' if len(x) >= 15 else 'short')
```

```python
df['name_result'] = df['name'].apply(lambda x : 'long' if len(x) >= 15 else 'short')
```

```python
df.head()
```

### 3. survived 1이면 생존, 0이면 사망, survived column update

```python
df['survived'].apply(lambda x : '생존' if x == 1 else '사망')
```

```python
df['survived'] = df['survived'].apply(lambda x : '생존' if x == 1 else '사망')
```

```python
df.head()
```

### 4. age 10대, 20대, 30대, 40대, 50대, 60대 이상으로 column update

```python
def age_result(x) :
    if 0 < x < 10 :
        return '유아'
    elif 10 <= x < 20 :
        return '10대'
    elif 20 <= x < 30 :
        return '20대'
    elif 30 <= x < 40 :
        return '30대'
    elif 40 <= x < 50 :
        return '40대'
    elif 50 <= x < 60 :
        return '50대'
    elif 60 <= x :
        return '60대 이상'
    else :
        return "확인불가"
```

```python
df['age'] = df['age'].apply(age_result)
df.head()
```

### 5. name에 Mr, Dr를 포함하고 있으면 '남자', Miss를 포함하고 있으면 '여자', 나머지 '기타'<br>파생변수 name_result2를 만들고, 각각을 카운트 남자 몇명, 여자 몇명

```python
def name_select(x) :
    if 'Mr.' in x or 'Dr.' in x:
        return '남자'
    elif 'Miss.' in x or 'Mrs.' in x:
        return '여자'
    else :
        return '확인불가'
```

```python
df['name_result2'] = df['name'].apply(name_select)
```
