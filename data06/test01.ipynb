```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
corr = df.corr() # 피어쓴 상관계수
corr
```

```python
cols = ['survived', 'pclass', 'age', 'sibsp', 'parch', 'fare', 'adult_male', 'alone']
```

```python
sns.set(font_scale=1.5)
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
sns.heatmap(corr.values,
           cbar = True,
           annot = True,
           square = True,
           fmt = '.2f', # 소수점
           xticklabels = cols,
           yticklabels = cols)
```

```python
sns.pairplot(df, height=3.5)
plt.show
```
