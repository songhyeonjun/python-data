```python
import pandas as pd
```

```python
df = pd.read_csv('folium-data.csv', encoding='utf-8')
df.head()
```

```python
del df['no']
df.head()
```

```python
df.set_index('자치구', inplace=True)
df.head()
```

```python
map = folium.Map(location=[37.56959058325873, 126.98314009994974],
                zoom_start=10)
map.choropleth(geo_data=geo_str,
              data=df['대기업'],
              columns=[df.index,df['대기업']],
              fill_color='PuRd',
              key_on = 'feature.id')
map
```

```python
import warnings
warnings.filterwarnings(action='ignore')
```

```python
map = folium.Map(location=[37.57, 126.98], 
                zoom_start=10
                )
map.choropleth(geo_data=geo_str,
                      data= df['대기업'],
                      columns=[df.index, df['대기업']],
                      fill_color='PuRd',
                      key_on='feature.id'
              )


for n in df.index: #서클 마커
    folium.CircleMarker([df['lat'][n], df['lng'][n]], 
                        radius = df['대기업'][n]*0.1, 
                        color='#3186cc', fill_color='#3186cc', fill=True).add_to(map)
map
```

```python
# cctv파일 열어서 2개 이상의 항목으로 folium으로 그리기
df2 = pd.read_csv('crime_seoul_location.csv')
df2.head()
```

```python
map2 = folium.Map(location=[37.56, 126.98],
                zoom_start=10)
map2.choropleth(geo_data = geo_str,
               data = df2['폭력 발생'],
               columns = [df2.index, df2['폭력 발생']],
               fill_color='PuRd',
               key_on='feature.id')
map2
```

```python
map2 = folium.Map(location=[37.56, 126.98],
                zoom_start=10)
map2.choropleth(geo_data = geo_str,
               data = df2['폭력 발생'],
               columns = [df2.index, df2['폭력 발생']],
               fill_color='PuRd',
               key_on='feature.id')

for n in df2.index: #서클 마커
    folium.CircleMarker([df2['lat'][n], df2['lng'][n]], 
                        radius = df2['폭력 발생'][n]*0.01, 
                        color='#3186cc', fill_color='#3186cc', fill=True).add_to(map2)
map2
```
