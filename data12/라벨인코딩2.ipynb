```python
# 1. beer_servings를 3개의 group으로 만들기
# 2. 1번의 결과를 beer_cat변수로 추가
# 3. LeberEncoding 2가지 방식으로
# 4. 원핫인코딩 df전체
# 5. 원핫인코딩 특정 컬럼(1개 또는 그 이상)을 선택해서
from sklearn.preprocessing import LabelEncoder
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('drinks.csv')
df
```

```python
a= df['beer_servings'].groupby(df['continent'])
a.sum()
```

```python
b_list= []
for i in df['beer_servings']:
    if i >= 300 :
        b_list.append('300++')
    elif i >= 150 and i < 300:
        b_list.append('150++')
    else :
        b_list.append('0++')
b_list
```

```python
df['beer_cat'] = b_list
df
```
