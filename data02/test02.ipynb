```python
import pandas as pd
import numpy as np
import random
```

```python
# 1. df로 읽어들이고
df = pd.read_csv('scientists.csv',  encoding='utf8')
df
```

```python
# 2. df의 전체 데이터셋 확인
df.dtypes
```

```python
# 3. df의 컬럼리스트를 확인
df.columns
```

```python
# 4. df의 값을 확인
df
```

```python
# 5. df의 맨 마지막에 price항목을 추가(값은 임의대로 넣음 int)]
price_list = []
for i in range(8):
    price_list.append(random.randint(100, 10000))
df['price'] =  price_list
df['price']
```

```python
# 6. df의 맨 마지막에 percent항목을 추가(값의 임의대로 넣음. float)
percent_list = []
for i in range(8):
    percent_list.append(float(random.randint(1.0, 100.0)))
df['percent'] =  percent_list
df['percent']
```

```python
df
```

```python
# 7. df의 맨마지막에 sale항목을 추가(price * percent)
df['sale'] = df['price'] - ( df['price'] * (df['percent'] / 100))
df
```
